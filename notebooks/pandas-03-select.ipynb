## How do I select a subset of a DataFrame
https://web.archive.org/web/20240318204924id_/https://pandas.pydata.org/pandas-docs/stable/getting_started/intro_tutorials/03_subset_data.html

```clojure
;; pip install pandas
(import [pandas :as pd])
nil
```

```clojure
(def titanic (pd/read_csv "https://web.archive.org/web/20230319165431id_/https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/titanic.csv"))
```

```clojure
titanic
```

```clojure
(.head titanic)
```

```clojure
(.tail titanic)
```

```clojure
(.info titanic)
```

```clojure
(def ages (aget titanic "Age"))
```

```clojure
ages
```

```clojure
(.-shape ages)
```

```clojure
(aget titanic ["Age" "Sex"])
```

```clojure
(def above-35 (aget titanic (> (aget titanic "Age") 35)))
```

```clojure
(.head above-35)
```

```clojure
(.-shape above-35)
```

```clojure
(> (aget titanic "Age") 35)
```

```clojure
(def class-23 (aget titanic (.isin (aget titanic "Pclass") [2 3])))
```

```clojure
(.head class-23)
```

```clojure
(operator/or (aget titanic "Pclass") (aget titanic "Pclass"))
```

```clojure
(def age-no-na (aget titanic (.notna (aget titanic "Age"))))
```

```clojure
(.-shape age-no-na)
```

```clojure
(aget (.loc titanic) (> (aget titanic "Age") 35)  "Name")
```

```clojure
(aget (.iloc titanic) #py ((slice 9 25) (slice 2 5)))
```

```clojure
(aset (.iloc titanic) #py ((slice 0 3) 3) "anonymous")
```
